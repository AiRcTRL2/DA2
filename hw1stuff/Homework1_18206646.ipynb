# 3. Exploring relationships between feature pairs

```python
# first we want to convert "Good" and "Bad" target feature to numerical values "0" and "1". We will create a copy 
# of the data frame to do this as we may not want to lose our data encoded as strings.

copyDataFrame = ctDataCleaned

copyDataFrame['RiskPerformance'].replace({'Good': 0, 'Bad': 1},inplace=True)
copyDataFrame['RiskPerformance'] = copyDataFrame['RiskPerformance'].astype("int64",copy=False)
```

```python
# Correlation matrix using code found on https://stanford.edu/~mwaskom/software/seaborn/examples/many_pairwise_correlations.html
sns.set(style="white")

# Select columns containing continuous data
continuous_columns = copyDataFrame[[  "RiskPerformance",\
                                      "ExternalRiskEstimate",\
                                      "PercentTradesNeverDelq",\
                                      "PercentInstallTrades",\
                                      "NetFractionRevolvingBurden",\
                                      "NetFractionInstallBurden",\
                                      "NumRevolvingTradesWBalance",\
                                      "NumInstallTradesWBalance",\
                                      "PercentTradesWBalance",\
                                      "MSinceOldestTradeOpen",\
                                      "MSinceMostRecentTradeOpen",\
                                      "AverageMInFile",\
                                      "NumSatisfactoryTrades",\
                                      "NumTrades60Ever2DerogPubRec",\
                                      "NumTrades90Ever2DerogPubRec",\
                                      "MSinceMostRecentDelq",\
                                      "MaxDelq2PublicRecLast12M",\
                                      "MaxDelqEver",\
                                      "NumTotalTrades",\
                                      "NumTradesOpeninLast12M",\
                                      "MaxDelq2PublicRecLast12M",\
                                      "NumInqLast6Mexcl7days",\
                                      "NumTradesOpeninLast12M",\
                                      "MaxDelq2PublicRecLast12M",\
                                      "NumInqLast6Mexcl7days"]].columns

# Calculate correlation of all pairs of continuous features
corr = copyDataFrame[continuous_columns].corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# set font size
sns.set(font_scale=1.2)

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(50, 15))

# align
f.autofmt_xdate()

# Generate a custom colormap - blue and red
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, annot_kws={"size": 12}, mask=mask, cmap=cmap, vmax=1, vmin=-1,
            square=True, xticklabels=True, yticklabels=True,
            linewidths=.7, cbar_kws={"shrink": .5}, ax=ax)
plt.yticks(rotation = 0)
plt.xticks(rotation = 70)
```

### The matrix above pointed out some interesting correlations. The reason for plotting this matrix was to further identify possible feature pairs to plot for further investigation.

- ExternalRiskEstimate correlates positively with MSinceMostRecentDelq, MSinceOldestTradeOpen, PercentTradesNeverDelq and AverageMInFile. The particularly interesting one here is that, based on this, it seems as time grows so does our external risk estimate.
- The percentage of trades that were never delinquent is moderately correlated with the number of satisfactory trades. I think I would have expected this correlation to be higher, as satisfactory trades are likely those which are not delinquent (generally). 
- Number of Total trades correlates positively with the Number of revolving trades with balance. This would imply that accounts that utilise this feature (of being extended credit) are more likely to trade.
- The number of satisfactory trades is extremely positively correlated with the number of trades. This would imply the more frequent you trade, the more likely your trades would be tagged as satisfactory. Interestingly, the number of total trades only moderately correlates with PercentTradesNeverDelq. Furthermore, this means that delinquency would not directly cause a trade to be unsatisfactory -- otherwise these correlations might be closer together.
- RiskPerformance was not necessarily plotted here to determine a strong correlation, but to be used as an indicator for the pairs to plot so we can further determine whether there is a relationship between a feature and the target outcome.

# 4. Transformation

Based on our above analysis, there are some decisions I have decided to make with regard to transforming our features:

- 1. Normalise the NumInqLast6Mexcl7days as this appears to be one of our best predictors of the target outcome and could be used in a ML model. This feature could be used in combination with a weighting to determine the target feature.
